```python
import librosa
import librosa.display
import IPython.display as display

path = "/content/Data/genres_original/classical/classical.00006.wav"
display.Audio(path)
```

```python
import numpy as np

y, sr = librosa.load(path)
S = librosa.feature.melspectrogram(y, sr=sr)
from PIL import Image
im =Image.fromarray(S).convert('L')
im
```

```python
im.save("/content/sp.tiff")

#from tiff to wav 
m = Image.open("/content/sp.tiff") 
import numpy as np
img=np.array(im)
print(img.dtype)
wav=librosa.feature.inverse.mel_to_audio(img.astype(np.float64))
```
